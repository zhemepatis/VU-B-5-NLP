
#### 1. Pateiktam sakiniui pateikite priklausomybių gramatikos medį, pakeisdami jo fono ir teksto spalvas (spalvas parinkite savo nuožiūra). Atspausdinkite mazgus ir priklausomybių žymas.

```python
import spacy
from spacy import displacy

text = 'The students exam scores were all fantastic'
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)

opt = {
    "bg": "#333",
    "color": "#fff"
}
displacy.render(doc, style="dep", jupyter=True, options=opt)
```

```python
opt = {
    "bg": "#592250",
    "color": "#F2B544",
    "compact": True,
    "add_lemma": True
}
displacy.render(doc, style="dep", jupyter=True, options=opt)
```

#### 4. Pateiktam sakiniui pateikite priklausomybių gramatikos medį, pakeisdami jo fono ir teksto spalvas (spalvas parinkite savo nuožiūra). Atskiroje eilutėje po sakinio tokenais atspausdinkite lemas. Atspausdinkite mazgus ir priklausomybių žymas.¶

```python
nlp = spacy.load("lt_core_news_sm")
text = "Žmogus be mokslo kaip namai be langų"
doc = nlp(text)

opt = {
    "bg": "#333",
    "color": "#fff",
    "compact": True,
    "add_lemma": True
}
displacy.render(doc, style="dep", jupyter=True, options=opt)
for token in doc:
    print(f"{token.text:{15}} {token.dep_:{15}} {token.head.text:{15}} {token.pos_:{15}} {[child for child in token.children]}")
```

#### 5. Iš duoto tekstinio failo "Tekstas.txt" atrinkite tik tuos sakinius, kuriuose yra veiksmažodžiai. Kiekvienam sakiniui, kuriame yra veiksmažodis, pavaizduokite gramatinį priklausomybių medį (dependency tree).

```python
with open("data/06_data.txt") as file:
    text = file.read()

nlp = spacy.load('en_core_web_sm')
doc = nlp(text)

target_sentences = []
for sentence in doc.sents:
    for word in sentence:
        if word.pos_ == "VERB":
            target_sentences.append(sentence)
            break

for sentence in target_sentences:
    sentence_doc = nlp(sentence.__str__().replace("\n", "").replace(".", ""))
    opt = {
    "bg": "#333",
    "color": "#fff",
    "compact": True,
    "add_lemma": True
    }
    displacy.render(sentence_doc, style="dep", jupyter=True, options=opt)

```

#### 6. Nustatykite kiekvieno sakinio galvą (angl. head) ir jos vaikus (angl. children)

```python
nlp = spacy.load("lt_core_news_sm")
text = "Vaikai žaidžia parke. Jis skaito naują knygą. Saulė šviečia ryškiai."
doc = nlp(text)

for sent in doc.sents:
    sent_doc = nlp(sent.__str__().replace("\n", "").replace(".", ""))

    opt = {
    "bg": "#333",
    "color": "#fff",
    "compact": True,
    "add_lemma": True
    }
    displacy.render(sent_doc, style="dep", jupyter=True, options=opt)

    result = []
    for token in sent_doc:
        result.append(f"{token.head.text:{15}} {[child for child in token.head.children]}")

    head_set = set(result)
    for item in head_set:
        print(item)
```
